# Import Libararies

```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
daterange = pd.date_range('01/02/2015', '31/12/2021')
da_lmp = pd.read_csv("da_lmp/20150101damlbmp_zone.csv")
rt_lmp = pd.read_csv("rt_lmp/20150101rtlbmp_zone.csv")
load_frcstd = pd.read_csv("load_forecast/20150101isolf.csv").iloc[:24,:]
for single_date in daterange:
    da_lmp = pd.concat([da_lmp, pd.read_csv("da_lmp/"+single_date.strftime("%Y%m%d")+"damlbmp_zone.csv")])
    rt_lmp = pd.concat([rt_lmp, pd.read_csv("rt_lmp/"+single_date.strftime("%Y%m%d")+"rtlbmp_zone.csv")])
    load_frcstd = pd.concat([load_frcstd, pd.read_csv("load_forecast/"+single_date.strftime("%Y%m%d")+"isolf.csv").iloc[:24,:]])
```

```python
zones = ['CAPITL', 'CENTRL', 'DUNWOD', 'GENESE', 'HUD VL', 'LONGIL',
       'MHK VL', 'MILLWD', 'N.Y.C.', 'NORTH', 'WEST']
da_lmp_zones_list = []
rt_lmp_zones_list = []
datetime_index = pd.date_range('01/01/2015', '31/12/2021 23:00:00',
                               freq='H').tz_localize("EST").tz_convert('UTC')

for i in zones:
    da_lmp_zones_list.append(da_lmp['LBMP ($/MWHr)'][da_lmp['Name'] == i].values)
    rt_lmp_zones_list.append(rt_lmp['LBMP ($/MWHr)'][rt_lmp['Name'] == i].values)
    
da_lmp_zones_df = pd.DataFrame(np.array(da_lmp_zones_list).T, columns = zones)
da_lmp_zones_df.index = datetime_index
rt_lmp_zones_df = pd.DataFrame(np.array(rt_lmp_zones_list).T, columns = zones)
rt_lmp_zones_df.index = datetime_index
load_frcstd_df = load_frcstd.iloc[:,1:-1].reset_index(drop=True)
load_frcstd_df.columns = zones
load_frcstd_df.index = datetime_index

```
