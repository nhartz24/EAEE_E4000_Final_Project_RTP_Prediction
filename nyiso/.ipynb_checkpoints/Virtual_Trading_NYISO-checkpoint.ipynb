# Read and Preprocess Data

```python
nyiso = pd.read_csv('full_nyiso.csv', index_col=0)
nyiso.index = pd.to_datetime(nyiso.index)
nyiso = nyiso.fillna(0)
```

```python
nyiso_zone_dict = {}
for idx, zone in enumerate(nyiso.zone.unique()):  
    nyiso_zone_dict[zone] = nyiso[nyiso.zone == nyiso.zone.unique()[idx]].iloc[:,1:]
    first_column = nyiso_zone_dict[zone].pop('lmp_spread')
    nyiso_zone_dict[zone].insert(0, 'lmp_spread', first_column)
    nyiso_zone_dict[zone]['Time'] = nyiso_zone_dict[zone].index.hour
    one_hot = pd.get_dummies(nyiso_zone_dict[zone]['Time'])
    nyiso_zone_dict[zone] = nyiso_zone_dict[zone].join(one_hot).drop(['Time'],axis=1)
    nyiso_zone_dict[zone].reset_index(inplace=True, drop=True)
    nyiso_zone_dict[zone].drop(nyiso_zone_dict[zone].index[[7392, 7394, 16130, 16132, 24868,
                                                            24870, 33606, 33608, 42512, 42514]], inplace=True)
#     nyiso_zone_dict[zone].set_index('Time Stamp', inplace=True, drop=True)
#     nyiso_zone_dict[zone].index = pd.to_datetime(nyiso_zone_dict[zone].index)

```

```python
nyiso_pred_zone_dict = {}
for idx, zone in enumerate(nyiso.zone.unique()):  
    nyiso_pred_zone_dict[zone] = pd.read_csv('nyiso_predictions_'+zone+'_19to21.csv', index_col=0)
```

# Optimization Formulation

Optimization Formulation:

$$ max \sum\sum(z^{I}*spread^{I} + z^{D}*-spread^{D}) $$

$$ s.t. \sum\sum(z^{I} + z^{D}) < B $$

$$ B = 50 $$

We have 11 zones, 24 hours a day

Lets try to do one day

```python
# preparing INC and DEC offers based on prediction values
I_pred_zone_dict = copy.deepcopy(nyiso_pred_zone_dict)
D_pred_zone_dict = copy.deepcopy(nyiso_pred_zone_dict)

I_pred_zone_dict[zone][I_pred_zone_dict[zone] < 0] = 0
D_pred_zone_dict[zone][D_pred_zone_dict[zone] > 0] = 0
```

```python
# preparing INC and DEC offers based on actual values
I_zone_dict = copy.deepcopy(nyiso_zone_dict)
D_zone_dict = copy.deepcopy(nyiso_zone_dict)

I_zone_dict[zone][I_zone_dict[zone] < 0] = 0
D_zone_dict[zone][D_zone_dict[zone] > 0] = 0
```

## Run Optimization for only one day

```python
# Run optimization based on spread predictions
z_i = cp.Variable((24,11), boolean = True)
z_d = cp.Variable((24,11), boolean = True)

profit = 0
for h in range(24):
    for n in range(11):
        profit += (z_i[h,n]*I_pred_zone_dict[nyiso.zone.unique()[n]].iloc[h]
                 - z_d[h,n]*D_pred_zone_dict[nyiso.zone.unique()[n]].iloc[h])

budget = 0
for h in range(24):
    for n in range(11):
        budget += z_i[h,n] + z_d[h,n]


objective = cp.Maximize(profit)
constraints = [budget <= 50]

prob = cp.Problem(objective, constraints)

prob.solve()
```

```python
# Evaluate the optimization by summing the profit based on the actual spreads
actual_profit = 0
for h in range(24):
    for n in range(11):
        actual_profit += (z_i.value[h,n]*I_zone_dict[nyiso.zone.unique()[n]].iloc[26304+h,0]
                         -z_d.value[h,n]*D_zone_dict[nyiso.zone.unique()[n]].iloc[26304+h,0])
actual_profit
```

## Set Budget

```python
B = 30
```

## Run Optimization for 2019 year

```python
# Run optimization based on spread predictions

profit_history = []
actual_profit_history = []

for day in range(365):
    
    z_i = cp.Variable((24,11), boolean = True)
    z_d = cp.Variable((24,11), boolean = True)

    profit = 0
    for h in range(24):
        for n in range(11):
            profit += (z_i[h,n]*I_pred_zone_dict[nyiso.zone.unique()[n]].iloc[day*24 + h]
                     - z_d[h,n]*D_pred_zone_dict[nyiso.zone.unique()[n]].iloc[day*24 + h])

    budget = 0
    for h in range(24):
        for n in range(11):
            budget += z_i[h,n] + z_d[h,n]


    objective = cp.Maximize(profit)
    constraints = [budget <= B]

    prob = cp.Problem(objective, constraints)

    profit_history.append(prob.solve())

    # Evaluate the optimization by summing the profit based on the actual spreads
    actual_profit = 0
    for h in range(24):
        for n in range(11):
            actual_profit += (z_i.value[h,n]*I_zone_dict[nyiso.zone.unique()[n]].iloc[17520 + day*24 + h,0]
                             -z_d.value[h,n]*D_zone_dict[nyiso.zone.unique()[n]].iloc[17520 + day*24 + h,0])
            
    actual_profit_history.append(actual_profit)
    
    if day%20 == 1:
        print(day)
```

```python
df = pd.DataFrame(data={"profit_history": profit_history,
                        "actual_profit_history": actual_profit_history})
df.to_csv("profit_history_2019_"+str(B)+".csv", sep=',',index=False)
```

```python
actual_profit_cumulative = [actual_profit_history[0]]
for i in range(1,len(actual_profit_history)):
    actual_profit_cumulative.append(sum(actual_profit_history[:i+1]))
```

```python
df = pd.DataFrame(data={"cumulative_profit": actual_profit_cumulative})
df.to_csv("cumulative_profit_2019_"+str(B)+".csv", sep=',',index=False)
```
